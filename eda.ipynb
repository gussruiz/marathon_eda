### Ultra-Marathon Exploratory Data Analysis

```python
import pandas as pd
import seaborn as sns
```

```python
## import and cleaning
```

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES")
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.dtypes
```

```python
# Getting only USA Races, 50k or 50mi in 2020
```

```python
df[df['Event distance/length'].isin(['50km', '50mi', '50Km'])]
```

```python
df[(df['Event distance/length'].isin(['50km', '50mi', '50Km'])) & (df['Year of event'] == 2020)]
```

```python
df[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']['Event name'] \
.str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA'] 
```

```python
df[(df['Event distance/length'].isin(['50km', '50mi', '50Km'])) & (df['Year of event'] == 2020) & \
    (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2 = df[(df['Event distance/length'].isin(['50km', '50mi', '50Km'])) & (df['Year of event'] == 2020) & \
    (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head()
```

```python
df2.shape
```

```python
# continue with the data cleaning
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
# cleaning athlete age
```

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
#remove h from athlete performance
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(1)
```

```python
# droping unnecessary columns
```

```python
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
df2.head(5)
```

```python
# cleaning null na values
```

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
#checking duplicates
```

```python
df2[df2.duplicated() == True]
```

```python
#resting index
```

```python
df2.reset_index(drop = True)
```

```python
#fixing types
```

```python
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int) 
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float) 
```

```python
df2.head()
```

```python
# renaming columns
```

```python

df2 = df2.rename (columns = {'Year of event': 'year',
                        'Event dates': 'race_day',
                        'Event name': 'race_name',
                        'Event distance/length': 'race_length',
                        'Event number of finishers': 'race_number_of_finishers', 
                        'Athlete performance': 'athlete_performance',
                        'Athlete gender': 'athlete_gender',
                        'Athlete average speed': 'athlete_average_speed', 
                        'Athlete ID': 'athlete_id'
})
```

```python
df2.columns
```

```python
# reordering columns
```

```python
df3 = df2[['race_day', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', \
    'athlete_gender', 'athlete_performance', 'athlete_average_speed',  'athlete_age']]
```

```python
df3.head()
```

```python
# charts and graphs
```

```python
sns.histplot(df3['race_length'])
```

```python
sns.histplot(df3, x = 'race_length', hue='athlete_gender')
```

```python
sns.displot(df3[df3['race_length'] == '50mi']['athlete_average_speed'])
```

```python
sns.violinplot(data = df3, x='race_length', y = 'athlete_average_speed', hue ='athlete_gender', \
    split = True, inner='quart', linewidth=1)
```

```python
sns.lmplot(data = df3, x='athlete_age', y = 'athlete_average_speed', hue = 'athlete_gender')
```

```python
# interesting analysis from the data (particular preference)
```

```python
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
# what ages has better speed
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed']\
    .agg(['mean', 'count']).sort_values('mean', ascending=False).query('count > 19').head(15)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed']\
    .agg(['mean', 'count']).sort_values('mean', ascending=True).query('count > 19').head(20)
```

```python
# the seasons affect the runner? 
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(\
    lambda x: 'Winter' if x > 11 \
    else 'Fall' if x > 8 \
    else 'Summer' if x > 5\
    else 'Spring' if x > 2\
    else 'Winter'
    )
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

```python
df3.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
